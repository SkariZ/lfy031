## Funktioner

En funktion är ett sätt gruppera ihop ett antal satser till en enhet som kan köras om på flera ställen. Kan förminska mängden kod drastiskt, då man slipper skriva samma sak flera gånger.

````python
# Notera indentering (antingen tab eller ett antal mellanslag) som visar vad som är funktionens implementering.
def say_hello(namn):
    print(f"Hej {namn}!")
    
def addera(a, b):
    return a + b

def multiplicera(a, b):
    return a*b

def pythagoras(a, b):
    return (a**2 + b**2)**0.5

#Även kombinera
def komb(a, b):
    return addera(a,b) + multiplicera(a,b)
````

````python
say_hello("Mio")

print(addera(3,8))

print(f"Pythagoras sats ger c={pythagoras(3, 4)} för a=3 och b=4")

komb(1, 1)
````

# Nyckelordsargument (keyword arguments)

En funktion i Python kan anropas med både s.k argument och keyword-argument.

````python
def f(a, b="standard b", c=42):
    print("a: ", a)
    print("b: ", b)
    print("c: ", c)
````

````python
f(0)
````

````python
f(0,1,2)
````

Detta kan användas t.ex när en funktion har många argument men där vi ofta inte vill specifiera alla när vi anropar.

### `*args, **kwargs`

Ibland kan det vara användbart att fånga upp extra argument och/eller nyckelordsargument till funktioner.

````python
def test(a, *args, **kwargs):
    print("Argument: ", args)
    print("Nyckelordsargument: ", kwargs)
````

````python
test(1, "hej", "hejsan", test=None, hejsan=False)

#test(test="hej", test2="")
````

`args` är en tupel medans `kwargs` är en dictionary. Deras speciella form i funktionsdefinitionen indikerar hur de kan användas:

````python
x = (1, "hej")
test(*x) # test(1, "hej")
````

````python
kwargs = {'nyckel': 'värde'}
test(1, **kwargs) # test(nyckel="värde")
````

Namnen måste i sig inte vara unika, men `args` och `kwargs` är standardbenämningen i Python

````python
def min_funktion(*positionella, **nyckelord):
    print("Positionella argument:", positionella)
    print("Nyckelordargument:", nyckelord)

# Anropa funktionen med olika antal argument
min_funktion(1, 2, 3, a=4, b=5)
````

## Läsa in filer med inbyggda Python bibliotek


Låt oss analysera väderdata från London insamlad mellan år 1979 och 2020 

https://www.ecad.eu/dailydata/index.php

Filnamn: `london_weather.csv`

csv-modulen: https://docs.python.org/3.6/library/csv.html

````python
import csv

# Öppna CSV-filen för läsning
with open('london_weather.csv', newline='') as csvfile:
    # Skapa en läsare för CSV-filen
    csv_reader = list(csv.reader(csvfile, delimiter=','))
````

````python
csv_reader
````

````python
data = []
senaste_dagarna = 5000 #Behöver kanske inte kolla för långt bak i tiden pga. beräkningshastighet m.m
for i, rad in enumerate(csv_reader):
    #Ta ut första listan samt de senaste dagarna
    if i > len(csv_reader) - senaste_dagarna or i == 0:
        data.append(rad) #Lägg till raden till data-listan med append
#Transformera till en numpy array        
data = np.array(data, dtype=object)
````

````python
data.shape
````

````python
data[0]
````

````python
data[1]
````

````python
data[-1]
````

````python
data[1].shape
````

````python
#Det finns endel strängar med tomma värden. Dvs. ''. Dessa bör man hantera på något lämpligt sätt.
data[4650, :]
````

````python
#Fyll saknade värden till 0
for i in range(1, len(data)):
    for j in range(1, len(data[i])):
        #Kollar om nuvarande element i rad i och kolumn j är en tom sträng. Om så, sätt till 0.
        if data[i][j] == '':
            data[i][j] = 0
data[4650, :]
````
