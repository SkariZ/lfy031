## Programmera med Python i textfiler

Ett Python-program är en samling av en eller flera textfiler med filändelsen `.py`

### program.py
```python
print("hej")
```

````python
def add_numbers(a, b):
    return a+b

def multiply(a, b):
    return a*b

def print_out(number):
    print(f"The number is {number}")
    
#Main Funktion. Vår "pipeline".
def main(number1, number2):
        n = add_numbers(a=number1, b=number2)
        n2 = multiply(a=n, b=number2)
        print_out(n2)

#Egentligen inte nödvändigt i detta fall, men hindrar python filen att köras vid importering.
if __name__=='__main__':
    
    #Man skulle kunna använda input() för att få argument via tangentbord, men här sätter vi dem själva
    a=7
    b=10
    
    #Kör main-funktion.
    main(number1=a, number2=b)
````

# Interaktiv programmering i terminal
På de flesta system finns även en interaktiv tolk (liknande Jupyter) direkt i terminalen.

```shell
ipython
```

`%run` och `%load` är till exempel två vanliga kommandon som tillhandahålls av IPython

1. **`%run`**:
   - `%run` används för att köra externa Python-skriptfiler från IPython-skalet.
   - Exempel: `%run min_script.py`.

2. **`%load`**:
   - `%load` används för att ladda innehållet från en extern Python-skriptfil direkt in i en cell i IPython.
   - Exempel: `%load example_script.py`.
